# Matplotlib Basics

Understanding the figures and axes

```python
import numpy as np
import matplotlib.pyplot as plt

#sample data set
x = np.linspace(0, 10, 100)
y = np.sin(x)

#Create a figure and axes
fig, ax = plt.subplots()

#create a line plot using the axes object
ax.plot(x, y)

#Display the plot
plt.show()
```

Customizing plots

```python
import numpy as np
import matplotlib.pyplot as plt

#sample dataset
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

#create a figure and axes
fig, ax = plt.subplots()

#create line plots with labels for the legend
ax.plot(x, y1, label='sin(x)')
ax.plot(x, y2, label='cos(x)')

#add titles and labels
ax.set_title('Customized plot')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

#add a legend
ax.legend()

#display the plot
plt.show()
```

Saving plots to files

```python
# sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

#create a figure and axes
fig, ax = plt.subplots()

#create line plot
ax.plot(x, y)

#add titles and labels
ax.set_title('A simple Line Plot')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

#save the plot to a file ie png format
fig.savefig('simple_line_plot.png')

plt.show()
```

# Plot Types

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

Bar Plots

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10,15,12,8]
plt.bar(categories, values)

plt.title('Bar plot')
plt.xlabel('Values')
plt.ylabel('Categories')

plt.show()
```

Histograms

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(size=1000)

plt.hist(data, bins=30)

plt.title('Histogram')

plt.show()
```

Scatter Plots

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.show()
```

Pie Charts

```python
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C', 'D']
sizes = [25,35,20,20]

plt.pie(sizes, labels=labels)
plt.show()
```

Box plot

It summarizes the distribution of a dataset by displaying its median, quartiles, and outliers using a box and whisker representation

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(size=1000)

plt.boxplot(data)

plt.title('Box plot')

plt.show()
```

Error Bars

Are used to show the variability of data by representing the uncertainity or error in the measurement

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 10)
y = np.sin(x)
errors = np.random.uniform(0.1, 0.3, size=10)

plt.errorbar(x, y, yerr=errors, fmt='o-')
plt.show()
```

# Customizing Plots

colors, markers, and Line Styles

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, color='red', linestyle='-', marker='o', label='sin(x)')
plt.plot(x, y2, color='blue', linestyle='--', marker='s', label='cos(x)')

plt.legend()
plt.show
```

Customizing axis ticks and labels

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

plt.xticks(np.arange(0, 11, step=1))
plt.yticks(np.arange(-1, 1.1, step=0.5))

plt.show()
```

Text and Annotations

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

plt.text(5, 0, 'Text at (5, 0)')
plt.annotate('Local max', xy=(1.57, 1), xytext=(2, 0.7),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

Customizing gridlines

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

plt.grid(color='gray', linestyle='--', linewidth=0.5)

plt.show()
```

Multiple plots on a single figure

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(x, y1)
ax1.set_title('sin(x)')

ax2.plot(x, y2)
ax2.set_title('cos(x)')

plt.show()
```

# Creating publication-quality figures

1. Customize plot aesthetics

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, color='red', linestyle='-', linewidth=2, label='sin(x)')
plt.plot(x, y2, color='blue', linestyle='--', linewidth=2, label='cos(x)')

plt.title('Sin(x) and Cos(x) Functions')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()

plt.show()
```

Adjust layout and sizing

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(6, 8), sharex=True)

ax1.plot(x, y1, color='red', linestyle='-', linewidth=2)
ax1.set_title('sin(x)')
ax1.set_ylabel('y-axis')

ax2.plot(x, y2, color='blue', linestyle='--', linewidth=2)
ax2.set_title('cos(x)')
ax2.set_xlabel('x-axis')
ax2.set_ylabel('y-axis')

plt.tight_layout()
plt.show()
```

 Export the figure in high resolution

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, color='red', linestyle='-', linewidth=2, label='sin(x)')
plt.plot(x, y2, color='blue', linestyle='--', linewidth=2, label='cos(x)')

plt.title('Sin(x) and Cos(x) Functions')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()

plt.savefig('sin_cos_functions.png', dpi=300)
```
